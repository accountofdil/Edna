```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRegressor
import pickle
```

```python
df.columns
```

```python
features = ['Price', 
            'Discount',
            'Inventory Level',
            'Promotion',
            'Competitor Pricing',
            'Category']
```

```python
target = 'Demand'
X = df[features].copy()
y = df[target]
```

```python
X
```

```python
y
```

```python
label_encoders = {}

categorical_cols = X.select_dtypes(include = 'object').columns

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le

label_encoders
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
xgb = XGBRegressor(objective = 'reg:squarederror', n_jobs = -1)
```

```python
xgb_params = {'n_estimators': [200, 300, 500],
              'max_depth': [3, 4, 6, 8],
              'learning_rate': [0.01, 0.05, 0.1],
              'subsample': [0.7, 0.8, 1.0],
              'colsample_bytree': [0.7, 0.8, 1.0],
              'min_child_weight': [1, 3, 5]}
```

```python
xgb_random_search = RandomizedSearchCV(estimator = xgb,
                                       param_distributions = xgb_params,
                                       n_iter = 25,
                                       scoring = 'neg_mean_absolute_error',
                                       cv = 3,
                                       verbose = 1,
                                       n_jobs = -1)
```

```python
xgb_random_search.fit(X_train, y_train)
```

```python
xgb_random_search.best_params_
```

```python
best_model = xgb_random_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
print(root_mean_squared_error(y_test, y_pred))
```

```python
feature_importance = pd.Series(best_model.feature_importances_,
                               index = X.columns).sort_values(ascending = False)
feature_importance
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
feature_importance.plot(kind = 'bar', color = 'black', ax = ax)
ax.set_title('Feature Importance', fontweight = 'bold', fontsize = 12)
ax.set_xlabel('Features', fontweight = 'bold')
ax.set_ylabel('Importance', fontweight = 'bold')
ax.grid(True, linestyle = '--', alpha = 0.3)
ax.tick_params(axis = 'x', rotation = 0)

plt.show()
```
